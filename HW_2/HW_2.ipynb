**3. Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до
36 и старше 36. Посчитайте количество футболистов в каждой категории.**

```python
def age_group(age_int):
    if age_int<=19:
      return 'До 20 лет'
    if 20<age_int<=30:
      return 'С 20 до 30 лет'
    if 30<age_int<=36:
      return 'С 30 до 36 лет'
    else:
      return 'Старше 36 лет'
```

```python
df_new['Age_group'] = df_new['Age'].apply(age_group)
```

```python
df_new['Age_group'].value_counts()
```
